```python
link2 = 'https://timetable.iate.obninsk.ru/'
```

```python
time_resp = requests.get(link2)
time_resp.url, time_resp.status_code
```

```python
with open('resp.html', 'w') as f:
    f.write(time_resp.text)
```

```python
time_resp.text
```

```python
text = time_resp.text.replace('\r', '').split('\n')
pprint.pp(text)
```

```python
ids = []
for i in range(len(text)):
    if '<div class="found-group">' in text[i]:
        l = text[i].replace(' ', '').split('"')
        group_id = l[-2].split('/')[-1]
        ids.append(int(group_id))
```

```python
len(ids)
```

```python
link3 = 'https://timetable.iate.obninsk.ru/group/'
time_resp = requests.get(link3+str(ids[54]))
with open('resp.html', 'w') as f:
    f.write(time_resp.text)
```
